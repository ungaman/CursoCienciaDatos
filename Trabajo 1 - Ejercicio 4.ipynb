Basado en estas observaciones preliminares, se escoge el método "read_csv" de Pandas para la importación de datos, ya que permite definir algunos parámetros como el caracter delimitador y el tratamiento de datos vacios.


```python
# Apertura de archivo e importación

# Para efectos de este ejercicio, el archivo CSV se encuentra en el mismo directorio que el notebook
dfDataCSV = pd.read_csv('Data - Riesgos.csv', sep=',', na_values=[''])
```

#### Requerimiento 2: 

a.  Mostrar estructura del Data Frame

```python
print('Dimensiones del DataFrame:')
print('Filas: ', len(dfDataCSV), ' Columnas: ', len(dfDataCSV.columns))
print('---')
print('Cantidad de elementos: ', dfDataCSV.size)
print('---')
print('Nombres de las Columnas: ')
print(dfDataCSV.columns)
print('---')
print('Tipos de datos de las Columnas:')
print(dfDataCSV.dtypes)
print('---')
print('Primeras 10 filas del DataFrame:')
print(dfDataCSV.head(10))
print('---')
print('Ultimas 10 filas del DataFrame:')
print(dfDataCSV.tail(10))
print('---')
print('Cantidad de datos "perdidos" o nulos: ', dfDataCSV.isna().sum().sum())
```

#### Requerimiento 3:
Copia de las 100 primeras filas hacia otro contenedor.

```python
dfNuevoSet = dfDataCSV.loc[0:99]
print('---')
print('Cantidad de filas copiadas: ', len(dfNuevoSet))
print('---')
print(dfNuevoSet)
```
